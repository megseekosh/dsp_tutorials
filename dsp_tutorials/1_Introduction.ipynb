## What is digital signal processing?<a name="whatis"></a> 

At its core, digital signal processing concerns the transformation of continuous signals, such as speech, into discrete representations that a computer can manage. This transformation from continuous (analog) to discrete (digital) signals must be done because computers simply can't store or handle measurements to an infinite number of decimal places. Some rounding must be performed. This is where the term **digital** comes from: the computer is storing discrete measurements in the form of *digits*.

When we take lot and lots of discrete measurements of a continuous signal like amplitude or time, we approximate a continuous function; however, crucially, even a large number of samples never *truly* results in the original, continuous signal. 

Before we explore the digitization of sound, let's practice reading an audio file into the computer and getting some basic information about it.

## Sampling rate<a name="sr"></a> 

Now that we know some basic ways to manipulate a sound file, we can return to the discussion about digitization. Afterall, that's what digital signal processing is all about!

There are two ways we can discretize continuous signals and they correspond to the two axes represented on our waveform: time and amplitude. 

First, we can discretize the x-axis or *time* of the function. This is the signal's **sampling rate**: how many times (per second) will we take a measurement? In speech and hearing sciences, we often take 41,000 samples/second for a sampling rate of 44.1kHz. 

There are several factors that go into a decision about sampling rate. First, unsurprisingly, since a higher sampling rate results in more measurements, audio signals encoded at higher sampling rates take up more space on a computer and are more computationally taxing. To see this, let's work with actual audio files.

Consider the load times for the following audio files that are approximately 1.5-hour and 14-hours in length:

```python
# different audio file load times
sr = 16000
%time
lb.load(path+'/long_files/vandam_public_45min.wav')
%time
long_file, sr = lb.load(path+'/long_files/BN32_010007.wav', sr=sr)
```

We can resample one of those files and see how much faster it is to process with a smaller sampling rate. There are lots of handy functions, like the Librosa one below, that make re-sampling audio files easy.

```python
long_file_8 = lb.resample(long_file, orig_sr=sr, target_sr=8000) # specify the original and target sampling rates
```

We can write the smaller file out to a directory and load it back in with the sound file now sampled at 8kHz::

```python
sf.write(path+'/long_files/vandam_full_8k.wav', long_file_8, 8000)
%time
long_file, sr = lb.load(path+'/long_files/vandam_full_8k.wav', sr=8000)
```

On the other hand, we don't want to *undersample* our signal. We need a sufficient number of samples to estimate the shape of our function and estimate components of the signal like its phase. Consider the following code:

```python
sr = 16000
t = np.arange(0,2,(1/sr))
a = np.sin(t*(2*np.pi)) 


fig, axs = plt.subplots(3,1,figsize=(8,8))
axs[0].plot(t,a,'darkgreen',lw=2) 
axs[0].set_title('16k samples/second') # # could fool you into thinking it's continuous!
axs[1].plot(t,a,'darkgreen',lw=2)
axs[1].set_title('The function sampled at 16k/second (green) with \n 5 samples/second overlaid on the first second (red)')  # much harder to tell the underlying shape of the function!
axs[1].plot(0,0,'or')
axs[1].plot(.25,1,'or')
axs[1].plot(.5,0,'or')
axs[1].plot(.75,-1,'or')
axs[1].plot(1,0,'or')
axs[2].plot(t,a,'darkgreen',lw=2)
axs[2].set_title('The function sampled at 16k/second (green) with \n 2 samples/second overlaid on the first second (red)') # the bare minimum number of samples needed to reconstruct the function
axs[2].plot(.25,1,'or')
axs[2].plot(.75,-1,'or')
plt.xlabel('Time')
fig.tight_layout()


for ax in axs.flat:
    ax.set(ylabel='Amplitude')
```

If we didn't have the underlying function to guide us, it would be much harder to "connect the dots" between samples in the 2 samples/second plot. 

Remember that sampling rate is the number of amplitude measurements we take per second. But the audio files in this tutorial come from very long recordings (up to 16 hours!). So, to save space and computation time, samples were only taken 16k times/second. 

Why does it matter how many samples we take per second? What happens when we specify the incorrect number of samples? Let's find out. 

```python
ipd.display(ipd.Audio(data=sound_file,rate=8000)) # under sample
ipd.display(ipd.Audio(data=sound_file,rate=44100)) # over sample
```

The current audio file has a 16k sampling rate. When we underspecify the sampling rate, the audio sounds grainy and the duration of the file has doubled. When we oversample the sampling rate, the pitch has increased and the duration of the file has decreased by approximately 60%.

Why does this occur? Recall that our audio object is just an array of measurements. No additional detail is provided to the computer. So how is the computer to know at what *frequency* those measurements were produced? How much time has elapsed between each sample? Over what time interval the samples were taken? We have to *specify* the distance between samples. 

When we *over-sampled* the recording, at 44.1kHz, we told the computer that our 23594-sample recording had 44100 samples per second. The computer condensed the distance between samples in the input, and the output was a shortened audio file and the perception of a higher pitch. 

The number of samples taken per second also determines what type of acoustic information can be transmitted. This value is known as the **Nyquist frequency**, or the sampling rate divided by two. The Nyquist frequency comes from the Nyquist-Shannon sampling theorem which tells us that to reliably reconstruct a sampled signal, we can only examine frequencies *below* one half the sampling rate. Or, in other words, we need to sample at twice the rate of the highest frequency that we hope to accurately represent in the sound. 

When we under-sample signals, we risk a phenomenon known as **aliasing**, so-called because unwanted, "aliased" frequencies appear in the signal. In practice, during analog-to-digital conversion in the computer's sound card, a filter is applied to a sound recording. Specifically, a *low-pass filter* (more on this later) is applied that attenuates all frequencies above the Nyquist frequency, or half the sampling rate (`.5*sr`). Alternatively, we can build filters to avoid aliasing into our signal processing programs. So we don't often have to worry about aliased frequencies appearing in our digital signals. But we should still understand the causes and consequences of aliasing, to be sure that we are avoiding it and to understand the motivation behind different sampling rates. Consider the following code:

```python
sr = 8000 # the Nyquist frequency is .5*sampling rate, or 4000Hz 
tone = lb.tone(2000,sr=sr,duration=1) # generate a pure tone well below Nyquist
ipd.display(ipd.Audio(data=tone,rate=sr)) # listen to it
```

Sounds like a pure tone! Next, we'll plot the tone's frequency by amplitude. Don't worry too much about the code below, we'll cover it in more detail in later tutorials:

```python
tone_fft = fft(tone) # y-values for plot
tone_freq = fftfreq(len(tone_fft),1/sr) # x-values for plot
(markers, stemlines, baseline) = plt.stem(abs(tone_freq), abs(tone_fft))
plt.setp(markers, marker='D', markersize=10, markeredgecolor="orange", color='indianred',markeredgewidth=2)
plt.setp(baseline, visible=False)
plt.setp(stemlines, linestyle="-", linewidth=4,color='indianred')
plt.xlabel('Freq (Hz)')
plt.ylabel('Amplitude')
plt.title('Frequency response of a single, 2000Hz tone')
plt.xlim(0,sr)
plt.axvspan(4000,8000,color='darkgreen',alpha=.3)
plt.annotate('Nyquist frequency',(3800,1000),size=15,color='darkgreen',weight='bold',rotation=90)
plt.ylim(0,6000)
```

The plot above looks good! There is one frequency in our signal, at 2000Hz. Just like we generated. Now let's generate a tone *at* Nyquist and see what happens:

```python
nyq_tone = lb.tone(4000,sr=sr,duration=1) # generate a pure tone at Nyquist (remember sr=8000)
ipd.display(ipd.Audio(data=nyq_tone,rate=sr)) # listen to it
```

That was...not a single pure tone. Let's plot the frequency response to see what's going on in the signal:

```python
nyq_tone_fft = fft(nyq_tone) # y-values for plot
nyq_tone_freq = fftfreq(len(nyq_tone_fft),1/sr) # x-values for plot
(markers, stemlines, baseline) = plt.stem(abs(nyq_tone_freq), abs(nyq_tone_fft))
plt.setp(markers, marker='D', markersize=6, markeredgecolor="cornflowerblue", color='navy',markeredgewidth=.5)
plt.setp(baseline, visible=False)
plt.setp(stemlines, linestyle="-", linewidth=4,color='navy')
plt.xlabel('Freq (Hz)')
plt.ylabel('Amplitude')
plt.title('Frequency response of a 4000Hz tone, sampled at 8kHz')
plt.xlim(0,sr)
plt.axvspan(4000,8000,color='darkgreen',alpha=.3)
```

There are now a *lot* of frequencies present in the signal---at 1000, 2000, 3000Hz and more. If you play the 4000Hz tone that we generated above, you can hear all of those different frequencies clashing together in the signal. What happened?

When we attempt to reconstruct a frequency above or even near Nyquist, those frequencies will be "folded" back onto the spectrum (folded from above Nyquist), as you can see in the frequency response for the 4000Hz tone above. This is why the Nyquist frequency is alternatively called the *folding frequency*. 

When these multiple frequencies appear in the signal, they interfere with one another (*destructive interference*, more on this in the Digital Filters tutorial) causing strange amplitude modulation effects. Specifically, when the frequencies' combined amplitudes meet, they cancel each other out. You can see some of this effect in the following waveforms:

```python
plt.plot(np.linspace(0,1,len(tone)),tone,color='indianred')
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.xlim(0,0.01)
plt.title('Waveform of single 2000Hz tone, appropriately sampled at 8000Hz')
```

```python
plt.plot(np.linspace(0,1,len(nyq_tone)), nyq_tone/.0000000006,color='navy')
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.xlim(0.4,0.42)
plt.ylim(-.005,.005)
plt.title('Waveform of "single" 4000Hz tone, inappropriately sampled at 8000Hz')
```

Once aliased frequencies are present in a sampled signal, it is impossible to remove them. There is simply no way to disentangle an aliased frequency from an actual frequency when they have the same value. 

You can play around with the code above to generate a simple tone well above Nyquist (at, say, 6000Hz), but you won't hear the same aliasing effect in the signal and you won't see it in the spectrum. You won't induce aliasing because the ready-made functions that we used to 1) generate (`lb.tone`) and 2) plot (`fft` and `fftfreq`) the tones have filters built in to protect against aliasing. But those functions don't filter for frequencies *at* Nyquist, so the 4000Hz tone results in aliasing in the signal. In the Digital Filter tutorial, we will discuss why even frequencies that approach Nyquist can result in aliasing, and will propose some solutions to the problem.

44.1kHz is the typical sampling rate used in speech and hearing sciences. Its Nyquist frequency is 22.05kHz, or just above the highest audible frequency to the human ear (approximately 20kHz if you're young and/or have excellent auditory health; many people can't hear frequencies this high, so don't worry if you cannot). However, there aren't any frequencies meaningful for human speech above about 10kHz, even in infant speech, so in reality a sampling rate of 22.05kHz, permitting the representation of frequencies up to 11.025kHz, is sufficient. 

For the 16kHz audio recording we've been playing with, we would only be able to capture speech frequencies up to 8kHz. For adult speech, this is acceptable since the spectral peak of the highest frequency fricatives, like [s],  hovers around 8kHz. But in children, the anterior cavity that generates the highest-frequency fricatives is much smaller and thus the frequencies generated are substantially higher (~9.5kHz). The recorders used to make the 16kHz audio recording we're working with were, ironically, designed to be used with children ([https://www.lena.org]). The creators should have spoken to a speech scientist who could have told them that the sampling rate was insufficient for the higher frequencies of child speech!

There might still be occasions when you want to record at sampling rates higher than even twice the audible range for human ears. **Ultrasonic frequencies** are frequencies above 20kHz and some animals, like bats, produce vocalizations in this range. Researchers have attached tiny audio recorders to bats to record their vocalizations (Kloepper et al., 2016), sampling the recordings at rates such as 250kHz. Even then, [bat vocalizations range from 9-200kHz](https://dnr.maryland.gov/wildlife/Pages/plants_wildlife/bats/batelocu.aspx), so not all frequencies would necessarily be captured. Now that you know about the Nyquist frequency, you understand why!

### A note about sampling rates and frequencies 

At this point, it's worth clarifying sampling rates versus signal frequencies, because both are measured in Hz which can be confusing. A signal's frequency (in Hz) represents how many times it repeats per second, or the number of cycles per second. So a 10Hz pure tone would have 10 cycles, or periods, per second:

```python
t = lb.tone(10,sr=sr,duration=1) 
plt.plot(np.linspace(0,1,len(t)), t, color='green',lw=2)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title('A 10Hz tone has 10 cycles/second')
```

Sampling rates refer to the number of samples taken each second. The red dots below represent a 10Hz sampling rate, taken from the same 10Hz sine wave (note the change in the x-axis), and the orange dots represent a 2Hz sampling rate: 

```python
fig, axs = plt.subplots(2,1,figsize=(8,8))
axs[0].plot(np.linspace(0,1,len(t)),t,'green')
axs[0].set_title('10Hz sampling rate over a 10Hz wave')
axs[0].plot(0.001,0,'or',markersize=10)
axs[0].plot(.01,0.6,'or',markersize=10)
axs[0].plot(0.02,0.96,'or',markersize=10)
axs[0].plot(0.03,0.96,'or',markersize=10)
axs[0].plot(0.04,.6,'or',markersize=10)
axs[0].plot(0.05,0,'or',markersize=10)
axs[0].plot(0.06,-.6,'or',markersize=10)
axs[0].plot(0.07,-.96,'or',markersize=10)
axs[0].plot(0.08,-.96,'or',markersize=10)
axs[0].plot(0.09,-.6,'or',markersize=10)
axs[1].plot(np.linspace(0,1,len(t)),t,'green')
axs[1].set_title('2Hz sampling rate over a 10Hz wave')
axs[1].plot(0.001,0,'or',markersize=10,c='darkorange')
axs[1].plot(.05,0,'or',markersize=10,c='darkorange')
plt.xlabel('Time (s)')
fig.tight_layout()

for ax in axs.flat:
    ax.set(ylabel='Amplitude',xlim=(0,0.1))
```

The important thing to remember is that Hz is a unit of measurement that reflects the frequency of a phenomenon. We express *signal* frequencies in Hz (number of cycles that repeat per second) and *sampling* frequencies in Hz (number of samples taken from each cycle). 

## Quantization<a name="quant"></a>  

We can also discretize our continuous signal along the y-axis or *amplitude* of the function. This is signal quantization: how detailed will our y-value (amplitude) measurements be?

The accuracy and detail of amplitude measurements depends on the number of **bits**, or units of information, used to encode the original audio information. If we encode with a higher number of bits, we will have a richer, more fine-grained representation of amplitude. The relationship between the number of bits and the resulting number of amplitude measures that we can capture is as follows:

```python
nbits = 4 # try plugging in different bit counts!
measures = 2**(nbits - 1)
print(nbits, 'bits results in', measures, 'amplitude measures')  
```

When we encode with a higher bit number, the resulting waveform is more detailed and the amplitude measures more exact. Consider the following sine waves:

```python
t = np.arange(1,10,(1/sr))
a = np.sin(t*(2*np.pi)) 

four = np.round(a * 2)*(1/2) # round to the .5 instead of integer
         # stretch to 2 numbers on the y-axis and then multiply by a fraction to reduce the y-axis values
         # without losing the quantization of the function
         # 4 amplitude measures = 2 bits
         # num amp measures = 2**(# bits)
sixteen = np.round(a * 16)*(1/16)        

ninesix = np.round(a * 96)*(1/96)  

fig, axs = plt.subplots(3,1,figsize=(8,8))
axs[0].plot(t,four,'tab:green')
axs[0].set_title('4 amplitude steps')
axs[1].plot(t,sixteen,'tab:orange')
axs[1].set_title('16 amplitude steps')
axs[2].plot(t,ninesix,'tab:red')
axs[2].set_title('96 amplitude steps')
plt.ylabel('Amplitude')
plt.xlabel('Time')
fig.tight_layout()

for ax in axs.flat:
    ax.set(ylabel='Amplitude')
```

The 96 amplitude step function could also fool you into thinking it's continuous! The signal to noise ratio, or **SNR**, is also higher when we use more bits to encode a signal. 

Just like with sampling rate, we can never have a *truly* continuous function represented digitally. Since we're always under-representing the original, continuous signal (to a certain degree), error is introduced and the result is noise between each amplitude step. That noise between steps is most obvious in the 4 amplitude step plot above, but it's important to remember that noise is an inherent component of quantization, and quantization is an inherent component of digital signals.

## Clipping<a name="clipping"></a>  

As a general rule of thumb, we want to maximize the amplitude of our speech recordings when we make them. Doing so allows us to maximize the signal to noise ratio during digital conversion. The maximum amplitude that we can encode during recording depends upon the bit depth of our analog-to-digital converter in the computer's sound card: a 16-bit converter permits higher amplitude encoding than an 8-bit converter. 

Consider each bit in the analog-to-digital converter like an empty bucket. If we want to maximize the volume of water transported, we should fill each bucket and not leave any empty. Similarly, during audio recording, we take full advantage of the number of bits available to us during digital conversion, whether that's 8-bit, 16-bit, etc., by maximizing the amplitude of our recording and therefore the amount of audio information transmitted.  

However, if we amplify our recording too much, the amplitude becomes *too* high and exceeds the range that our microphone can accept. The result is that portions of the signal are lost in a process called **clipping.** Clipping means that all measurements outside of a pre-determined range (the voltage sensitivity range of the machine's analog-to-digital converter (Johnson, 2012:58)), will be reduced to the most extreme amplitude values (-1 or 1). The result could be, for example, a sine wave transformed to a square wave form: 

```python
# sine wave with amplitudes outside of -1 to 1
sr = 41000
t = np.arange(0, 1, (1/sr))
sf = np.sin(t*(2*np.pi)*5)*2

plt.figure()
plt.axhline(y=1, color='red')
plt.axhline(y=-1, color='red')
plt.plot(t,sf)
plt.ylabel('Amplitude')
plt.xlabel('Time')
plt.title("High-amp sine wave")
plt.show()

# cut frequencies > abs(1) off
def clip_audio(audio):
    cut_a = []
    for s in audio:
        if s > 1:
            cut_a.append(1)
        elif s < -1:
            cut_a.append(-1)
        else:
            cut_a.append(s)
    return cut_a

cut_sf = clip_audio(sf)

# show resulting square wave
plt.figure(figsize=(6,2))
plt.axhline(y=1, color='red')
plt.axhline(y=-1, color='red')
plt.plot(t,cut_sf)
plt.ylabel('Amplitude')
plt.xlabel('Time')
plt.title("'Sine' wave after hard clipping")
plt.show()
```

We can synthesize a higher-frequency sine wave and clip it to hear the effect of clipping upon audio:

```python
sf_100 = np.sin(t*(2*np.pi)*100)*2
cut_sf_100 = clip_audio(sf_100)

# listen to the original and the clipped audio
ipd.display(ipd.Audio(data=sf_100,rate=sr))
ipd.display(ipd.Audio(data=cut_sf_100,rate=sr))
```

The clipped one does not sound great. 

There is no way to restore the original amplitude measurements once they are clipped. But you can ensure that the amplitude of your recording is within a range acceptable to your microphone by closely watching the waveform during recording and ensuring that it stays within the acceptable -1 to 1 range. If it falls outside of that range at any point, you should lower the gain meter (input into the system) on the microphone pre-amplifier. 

## Mono vs stereo signals<a name="mono"></a>  

The last topic that will be covered in this introductory tutorial is mono versus stereo audio signals. To illustrate these concepts, let's create a stereo file. I recommend wearing headphones to listen to these tracks so that you can hear how the signal is manipulated across your ears. 

As we saw above when we read the audio file into the computer, a mono audio signal is just a 1D array of amplitude measures. A stereo signal, as it turns out, is just a 2D array (a matrix) of amplitude measures, with one array representing the channel that will be directed to the left ear and one array representing the channel that will be directed to the right ear. Consequently, we can combine two 1D arrays (of the same length) together to create a stereo sound. The sound file that we previously loaded is mono, but we could change that. If a stereo sound is just a 2D array, we can add an "empty" array to the left channel, forcing the original sound to play from our right headphone.

First we create an array of 0s. Crucially this must be the same length as the sound file:

```python
# we'll load sound_file again
sound_file, sr = lb.load(path+'/files/Cychosz_105_740_1138.76_1139.83.wav')

e = np.linspace(0,0,len(sound_file)) # this function says we want an array over the interval 0-0 that is
                                     # the same length as the sound file
```

Then we create a 2D array or matrix combining the original sound file and the new "empty" array:

```python
r_ear = np.array([e,sound_file])

ipd.display(ipd.Audio(data=r_ear,rate=sr)) # plays out of the right headphone!

l_ear = np.array([sound_file,e])

ipd.display(ipd.Audio(data=l_ear,rate=sr)) # plays out of the left headphone!
```

More commonly we might want to combine two audio tracks; for example, from the two speakers in a sociolinguistic interview. To illustrate this, I'll read in a *different* mono audio file of a baby vocalizing and will create a 2D array as before:

```python
sound_file2, sr = lb.load(path+'/files/Cychosz_105_740_1799.98_1801.91.wav') 

stereo_sound = np.array([sound_file,sound_file2])

ipd.Audio(data=stereo_sound,rate=sr)
```

Now it sounds like there are babies are vocalizing to our left and right!

There are, of course, built-in functions to convert from stereo to mono in libraries such as Librosa:

```python
new_mono = lb.to_mono(stereo_sound)
new_mono.shape # this is indeed a 1D array
```

But it's also important to have a conceptual understanding of the shape of mono versus stereo signals, and now you do.
